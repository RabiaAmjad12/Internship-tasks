```python
import pandas as pd
```

```python
df= pd.read_csv("iris.csv")
df.head(150)
```

```python
# shape of the dataframe
df.shape
```

```python
# columns of the dataframe
df.columns
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
# histogram (sepal length distributin)
plt.figure(figsize=(8,6))
sns.histplot(data=df, x="SepalLengthCm",  kde=True, bins=20, color="red")
plt.title("Distribution of sepal length")
plt.xlabel("sepal length (cm)")
plt.ylabel("frequency")
plt.show()
```

```python
# BOX PLOT petal width
plt.figure(figsize=(8,6))
sns.boxplot(data=df, x="Species", y="PetalWidthCm", color="green")
plt.title("Box plot of petal width")
plt.xlabel("species")
plt.ylabel("petal width (cm)")
plt.show()
```

```python
# scatter plot sepal length vs petal length
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x="SepalLengthCm", y="PetalLengthCm", hue="Species", palette="Set1")
plt.title("scatter plot of sepal length vs petal length")
plt.xlabel("sepal length (cm)")
plt.ylabel("petal length (cm)")
plt.show()
```
